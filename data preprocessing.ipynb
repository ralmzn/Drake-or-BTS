```python
import pandas as pd
import numpy as np
```

```python
drake = pd.read_csv('data/drake_data.csv')
drake
```

```python
drake['artist'] = 'Drake'

```

```python
drake = drake[['lyrics', 'artist']]
```

```python
drake
```

```python
bts = pd.read_csv('data/bts.csv', sep=',', encoding='latin-1')
```

```python
bts['artist'] = 'BTS'
```

```python
bts = bts[['lyrics', 'artist']]
```

```python
combine = pd.concat([drake, bts])
```

```python
combine.dropna(inplace=True)
```

```python
combine.artist.value_counts()
```

```python
combine['is_drake'] = combine['artist'].map({'Drake':1, 'BTS':0})
```

```python
combine.drop('artist', axis=1, inplace=True)
```

```python
combine
```

```python
combine.to_csv('data/lyrics.csv', index=False)
```
